```python
import pandas as pd

# Replace this with the actual path to your CSV file or another data source
data_source_path = "/content/drive/MyDrive/Colab Notebooks/Sample_Workbook.xlsx"
output_master_path = "/content/drive/MyDrive/Colab Notebooks/Master_Workbook.csv"

# Ingest data from an external source (e.g., CSV file)
ingested_data = pd.read_excel(data_source_path)

# Additional data processing or manipulation steps can be added here if needed
# For example, you might filter, transform, or clean the ingested_data

# Create an empty DataFrame to store the consolidated data
consolidated_data = pd.DataFrame()

# Append the ingested data to the consolidated_data DataFrame
consolidated_data = consolidated_data.append(ingested_data, ignore_index=True)

# Write the consolidated data to a new Excel workbook
consolidated_data.to_csv(output_master_path, index=False)

print(f"Data from {data_source_path} ingested, consolidated, and saved to {output_master_path}")
```
